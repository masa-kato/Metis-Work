### Challenge 3.1

**Plot Domestic Total Gross Over Time**

```python
def plot_dom_tot_vs_time(df):
    '''
    Input: Pandas Movie Dataframe
    Output: Plot of Domestic Total Gross Over Time
    '''
    
    dom_tot = df['DomesticTotalGross'] 
    date = df['ReleaseDate']
    date = date.apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
    
    
    plt.figure(figsize=(12,8))
    plt.plot_date(date, dom_tot)
    plt.xlabel('Release Date',fontsize=18)
    plt.ylabel('Domenstic Total Gross',fontsize=18)
    plt.title('Domestic Total Gross Over Time',fontsize=20)
    plt.tick_params(axis='both', labelsize=12)
```

```python
plot_dom_tot_vs_time(df)
```

### Challenge 3.2

**Plot Runtime vs. Domestic Total Gross**

```python
def plot_runtime_vs_dom_tot(df):
    '''
    Input: Pandas Movie Dataframe
    Output: Plot of Domestic Total Gross Over Time
    '''
    
    dom_tot = df['DomesticTotalGross'] 
    runtime = df['Runtime']    
    
    plt.figure(figsize=(12,8))
    plt.scatter(runtime, dom_tot)
    plt.xlabel('Runtime',fontsize=18)
    plt.ylabel('Domestic Total Gross',fontsize=18)
    plt.title('Runtime vs Domestic Total Gross',fontsize=20)
    plt.tick_params(axis='both', labelsize=12)
```

```python
plot_runtime_vs_dom_tot(df)
```

### Challenge 3.3

**Group data by rating and find the average runtime and domestic total gross at each level of rating**

```python
def plot_grouped_rating(df):
    return df.groupby('Rating')[['Runtime','DomesticTotalGross']].mean()
```

```python
plot_grouped_rating(df)
```

### Challenge 3.4

**Make one figure with N subplots, and in each plot the release date vs the domestic total gross where N = the number of MPAA ratings**

```python
def plot_date_vs_dom_tot(df):
    plt.figure(1)
    plt.figure(figsize=(20,15))

    plt.subplot(221)
    plt.plot_date(df[df['Rating']=='G']['ReleaseDate'], df[df['Rating']=='G']['DomesticTotalGross'])
    plt.xlabel('Release Date',fontsize=14)
    plt.ylabel('Domenstic Total Gross',fontsize=14)
    plt.title('Rating: G',fontsize=16)
    
    plt.subplot(222)
    plt.plot_date(df[df['Rating']=='PG']['ReleaseDate'], df[df['Rating']=='PG']['DomesticTotalGross'])
    plt.xlabel('Release Date',fontsize=14)
    plt.ylabel('Domenstic Total Gross',fontsize=14)
    plt.title('Rating: PG',fontsize=16)    
    
    plt.subplot(223)
    plt.plot_date(df[df['Rating']=='PG-13']['ReleaseDate'], df[df['Rating']=='PG-13']['DomesticTotalGross'])
    plt.xlabel('Release Date',fontsize=14)
    plt.ylabel('Domenstic Total Gross',fontsize=14)
    plt.title('Rating: PG-13',fontsize=16)    
    
    plt.subplot(224)
    plt.plot_date(df[df['Rating']=='R']['ReleaseDate'], df[df['Rating']=='R']['DomesticTotalGross'])
    plt.xlabel('Release Date',fontsize=14)
    plt.ylabel('Domenstic Total Gross',fontsize=14)
    plt.title('Rating: R',fontsize=16)    
```

```python
plot_date_vs_dom_tot(df)
```

### Challenge 3.5

**What director in your dataset has the highest gross per movie?**

```python
def highest_director(df):
    temp = df.groupby('Director')['DomesticTotalGross'].mean()
    return temp[temp == temp.max()].index[0]
```

```python
highest_director(df)
```

### Challenge 3.6

**Bin your dataset into months and make a bar graph of the mean domestic total gross by month. Error bars will represent the standard error of the mean.**

**Title of graph should include: Mean Domestic Total Gross by Month in 2013**

**Topic for consideration: what is the correct formula for the standard error of the mean? Examine the error bars and see if they are "reasonable."**

```python
def plot_bar_graph(df):
    df = df.copy()
    df['ReleaseDate'] = df['ReleaseDate'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))

    grouped_mean = df.groupby(df['ReleaseDate'].map(lambda x: x.month))['DomesticTotalGross'].mean()    
    grouped_std = df.groupby(df['ReleaseDate'].map(lambda x: x.month))['DomesticTotalGross'].std()
    
    index = np.array(list(range(12)))
    labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    bar_width = 1
    
    plt.figure(figsize=(12,10))
    plt.bar(index, grouped_mean, width=bar_width,yerr=grouped_std,alpha=0.5)
    
    plt.xticks(index + bar_width/2, labels)
    plt.xlabel('Month',fontsize=18)
    plt.ylabel('Mean Domestic Total Gross',fontsize=18)
    plt.title('Mean Domestic Total Gross by Month',fontsize=24)
```

```python
plot_bar_graph(df)
```

```python
def plot_bar_graph_alt(df):
    df = df.copy()
    df['ReleaseDate'] = df['ReleaseDate'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
    grouped_mean = df.groupby(df['ReleaseDate'].map(lambda x: x.month))['DomesticTotalGross'].mean()    
    grouped_std = df.groupby(df['ReleaseDate'].map(lambda x: x.month))['DomesticTotalGross'].std()
    labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    
    
    plt.figure(figsize=(12,10))
    ax = sns.barplot(x=labels, y=grouped_mean,color='salmon',ci=68)
    plt.xlabel('Month',fontsize=18)
    plt.ylabel('Mean Domestic Total Gross',fontsize=18)
    plt.title('Mean Domestic Total Gross by Month',fontsize=24)
```

```python
plot_bar_graph_alt(df)
```
