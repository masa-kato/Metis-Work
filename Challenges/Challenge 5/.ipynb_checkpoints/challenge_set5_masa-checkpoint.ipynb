```python
# Import Relevant Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# Import Data
df = pd.read_csv('2013_movies.csv')
```

**Create Plotting Functions**

```python
def plot_predictions(y_pred, y_true):
    plt.figure(1)
    plt.rcParams['figure.figsize']=(14,10)
    plt.scatter(x=y_true,y=y_pred)
    
    # Prettify
    sns.plt.title('Prediction vs Actual',fontsize=26)
    sns.plt.xlabel('Actual Y', fontsize=20)
    sns.plt.ylabel('Predicted Y', fontsize=20)
    plt.tick_params(axis='both', labelsize=12)
```

```python
def plot_residuals(residuals):
    plt.figure(2)
    plt.rcParams['figure.figsize']=(14,10)
    plt.hist(x=residuals, bins=15)
    
    # Prettify
    sns.plt.title('Residual Distribution Plot',fontsize=26)
    sns.plt.xlabel('Residual', fontsize=20)
    sns.plt.ylabel('Counts', fontsize=20)
    plt.tick_params(axis='both', labelsize=12)
```

## Challenge 5.1

**Build a linear model that uses only a constant term (a column of ones) to predict a continuous outcome (like domestic total gross). How can you interpret the results of this model? What does it predict? Make a plot of predictions against actual outcome. Make a histogram of residuals. How are the residuals distributed?**

```python
def challenge_1(df):
    df = df.copy()
    model = LinearRegression()
    df['Constant'] = 1
    
    # Define Features & Target
    features = ['Constant']
    target = ['DomesticTotalGross']
    X = df[features]
    y = df[target]
    
    # Create Model
    model.fit(X,y)
    y_pred = model.predict(X)
    y_true = np.array(y)
    residuals = y_true - y_pred
    
    # Print Predictions
    print('Predictions of First Five Movies:')
    print(y_pred[:5]) 

    # Plots
    plot_predictions(y_pred, y_true)
    plot_residuals(residuals)
    
    return model
```

```python
challenge_1(df)
```

## Challenge 5.3

**Repeat the process of challenge 1, but add a categorical feature (like genre). You'll have to convert a column of text into a number of numerical columns ("dummy variables"). How can you interpret what's happening in the model?**



```python
def challenge_3(df):
    df = df.copy()
    model = LinearRegression()
    df['Constant'] = 1
    df['Budget'].fillna(df['Budget'].mean(), inplace=True)
    rating_dummy = pd.get_dummies(df['Rating'],prefix='Rating')
    df = pd.concat([df, rating_dummy], axis=1)

    # Define Features & Target
    features = ['Constant','Budget','Rating_PG-13','Rating_PG','Rating_G','Rating_R']
    target = ['DomesticTotalGross']
    X = df[features]
    y = df[target]
    
    # Create Model
    model.fit(X,y)
    y_pred = model.predict(X)
    y_true = np.array(y)
    residuals = y_true - y_pred
    
    # Print Predictions
    print('Predictions of First Five Movies:')
    print(y_pred[:5]) 

    # Plots
    plot_predictions(y_pred, y_true)
    plot_residuals(residuals)
    
    return model
```

## Challenge 5.4

**Enhance your model further by adding more features and/or transforming existing features. Think about how you build the model matrix and how to interpret what the model is doing.**

```python
def challenge_4(df):
    df = df.copy()
    model = LinearRegression()
    df['Constant'] = 1
    df['Budget'].fillna(df['Budget'].mean(), inplace=True)
    rating_dummy = pd.get_dummies(df['Rating'],prefix='Rating')
    df = pd.concat([df, rating_dummy], axis=1)

    # Define Features & Target
    features = ['Constant','Budget','Rating_PG-13','Rating_PG','Rating_G','Rating_R','Runtime']
    target = ['DomesticTotalGross']
    X = df[features]
    y = np.log(df[target])
    
    # Create Model
    model.fit(X,y)
    y_pred = model.predict(X)
    y_true = np.array(y)
    residuals = y_true - y_pred
    
    # Print Predictions
    print('Predictions of First Five Movies:')
    print(y_pred[:5]) 

    # Plots
    plot_predictions(y_pred, y_true)
    plot_residuals(residuals)
    
    return model
```

## Challenge 5.5

**Fitting and checking predictions on the exact same data set can be misleading. Divide your data into two sets: a training and a test set (roughly 75% training, 25% test is a fine split). Fit a model on the training set, check the predictions (by plotting versus actual values) in the test set.**

```python
def challenge_5(df):
    df = df.copy()
    model = LinearRegression()
    df['Constant'] = 1
    df['Budget'].fillna(df['Budget'].mean(), inplace=True)
    rating_dummy = pd.get_dummies(df['Rating'],prefix='Rating')
    df = pd.concat([df, rating_dummy], axis=1)

    # Define Features & Target
    features = ['Constant','Budget','Rating_PG-13','Rating_PG','Rating_G','Rating_R','Runtime']
    target = ['DomesticTotalGross']
    X = df[features]
    y = np.log(df[target])
    
    # Train Test Split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0)
    
    # Create Model
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    y_true = np.array(y_test)
    residuals = y_true - y_pred
    
    # Print Predictions
    print('Predictions of First Five Movies:')
    print(y_pred[:5]) 

    # Plots
    plot_predictions(y_pred, y_true)
    plot_residuals(residuals)
    
    return model
```

```python
challenge_5(df)
```
