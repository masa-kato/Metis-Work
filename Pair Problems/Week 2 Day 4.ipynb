```python
num_feat = list(range(1,101))
np.random.seed(0)
y = np.array(np.random.rand(200, 1))

r_2 = []
adj_r_2 = []
for feat in num_feat:
    np.random.seed(0)
    lin_reg = sm.OLS(np.log(y), np.array(np.random.rand(200, feat))).fit()
    r_2.append(lin_reg.rsquared)
    adj_r_2.append(lin_reg.rsquared_adj)

    
plt.figure(figsize = (10,8))
plt.plot(num_feat, r_2, label='R^2')
plt.plot(num_feat, adj_r_2, label='Adj R^2')
plt.legend()
plt.show()
```

```python
num_feat = list(range(1,101))
np.random.seed(0)
y = np.array(np.random.rand(200, 1))
np.random.seed(0)
X = np.array(np.random.rand(200,100))

r_2 = []
adj_r_2 = []
for feat in num_feat:
    lin_reg = sm.OLS(np.log(y), X).fit()
    r_2.append(lin_reg.rsquared)
    adj_r_2.append(lin_reg.rsquared_adj)
    X = np.delete(X, obj=max(num_feat)-feat ,axis=1)

r_2 = r_2[::-1]
adj_r_2 = adj_r_2[::-1]

plt.figure(figsize = (10,8))
plt.plot(num_feat, r_2, label='R^2')
plt.plot(num_feat, adj_r_2, label='Adj R^2')
plt.legend()
plt.show()
```
