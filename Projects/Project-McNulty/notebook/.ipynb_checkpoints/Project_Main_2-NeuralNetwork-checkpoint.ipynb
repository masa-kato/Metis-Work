```python
def sigmoid(z):
    '''
    Input:
        z = X * theta
    '''
    return 1 / (1 + np.exp(-z))

def relu(z):
    '''
    Input:
        z = X * theta
    '''    
    return np.maximum(0, z)

def softmax(z, m):
    '''
    Input:
        z = X * theta
    '''   
    z = np.copy(z)
    # Shift Values such that Max Number is 0 (For Numerical Stability)
    z -= np.max(z, axis=1).reshape(m,1)
    return np.exp(z) / np.sum(np.exp(z), axis=1, keepdims=True)
```

```python
class NeuralNetwork_MK(object):
    def __init__(self, input_size, hidden_size, output_size, model=None, activation=relu):
        # Set Layer Sizes
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        # Activation Function
        self.activation = activation
        
        # Initialize Dictionary of Bias/Weights
        self.bias = {}
        self.thetas = {}
        
        if model == None:
            # Initialize Bias to Zeros
            self.bias['B1'] = np.zeros(self.hidden_size)
            self.bias['B2'] = np.zeros(self.output_size)
            # Initialize Weights to Small Random Values
            self.thetas['Theta1'] = 1e-5 * np.random.randn(self.input_size, self.hidden_size)
            self.thetas['Theta2'] = 1e-5 * np.random.randn(self.hidden_size, self.output_size)
        
        else:
            # Initialize Bias to Input Model Values
            self.bias['B1'] = model['B1']
            self.bias['B2'] = model['B2']
            # Initialize Weights to Input Model Values
            self.thetas['Theta1'] = model['Theta1']
            self.thetas['Theta2'] = model['Theta2']
        
    def forward_propagation(self, X, y, lambda_reg):
        # Retrieve Number of Observations & Features
        m, n = X.shape
        # Retrieve Parameters
        theta1, bias1 = self.thetas['Theta1'], self.bias['B1']
        theta2, bias2 = self.thetas['Theta2'], self.bias['B2']
        
        # Layer 1
        self.a1 = np.copy(X)
        
        # Layer 2 (Hidden Layer)
        self.z2 = np.dot(self.a1, theta1) + bias1
        self.a2 = self.activation(self.z2)
        
        # Layer 3 (Output Layer)
        self.z3 = np.dot(self.a2, theta2) + bias2
        self.prob = softmax(self.z3, m)
        
        # If No Targets are Given Return the Scores
        if y == None:
            return self.z3
        
        # Cross Entropy Cost Function (with Regularization)
        log_prob = -np.log(self.prob)
        log_prob_target = log_prob[range(m), y]
        cost = np.sum(log_prob_target) / m 
        reg_cost = 0.5 * lambda_reg * (np.sum(theta1*theta1) + np.sum(theta2*theta2))
        total_cost = cost + reg_cost
        return total_cost

    def backward_propagation(self, X, y, lambda_reg):
        # Retrieve Values:
        prob = np.copy(self.prob)
        m, n = X.shape
        theta1, bias1 = self.thetas['Theta1'], self.bias['B1']
        theta2, bias2 = self.thetas['Theta2'], self.bias['B2']     
        # Initialize Gradient Dictionary
        grad_thetas = {}
        grad_bias = {}
        
        # Compute Gradient of Cross Entropy Cost Function
        gradient_cost = np.copy(prob)
        gradient_cost[range(m), y] = gradient_cost[range(m), y] - 1
        gradient_cost /= m
        
        # Compute Gradient of Weights, Bias in Hidden Layer
        grad_bias['B2'] = np.sum(gradient_cost, axis=0)
        grad_thetas['Theta2'] = np.dot(np.transpose(self.a2), gradient_cost) + lambda_reg * theta2
        
        # Compute Gradient of Hidden Layer
        gradient_hidden = np.dot(gradient_cost, np.transpose(theta2))
        gradient_hidden[self.z2 <= 0] = 0

        # Compute Gradient of Weights, Bias in Input Layer
        grad_thetas['Theta1'] = np.dot(np.transpose(self.a1), gradient_hidden) + lambda_reg * theta1
        grad_bias['B1'] = np.sum(gradient_hidden, axis=0)
        
        return grad_thetas, grad_bias
        
    def fit(self, X_train, y_train, X_test, y_test, batch_size=False, lambda_reg=0, num_epoch=100, 
            alpha_learn=1e-3, alpha_decay=0.95, print_costs=False, update_method='sgd',momentum=0):
        # Retrieve Values:
        X = np.array(X_train)
        y = np.array(y_train)
        X_test = np.array(X_test)
        y_test = np.array(y_test)
        m, n = X.shape
        
        # Initialize Variables:
        train_costs = []
        test_costs = []
        training_accuracys = []
        test_accuracys = []
        velocity = {}
        
        best_accuracy = 0
        best_parameters = {}
        
        # Set Number of Iterations
        if batch_size != False:
            iterations_per_epoch = m / batch_size
        else:
            iterations_per_epoch = 1
        iterations = int(num_epoch * iterations_per_epoch)
        
        for iteration in range(iterations):
            if iteration%10 == 0:
                print('Iteration {}'.format(iteration))
            
            if batch_size == False:
                # Set Batch for Batch Gradient Descent
                X_mini = np.copy(X)
                y_mini = np.copy(y)
            else:    
                # Set Minibatches for Mini-Batch Gradient Descent
                minibatch_index = np.random.choice(range(m), batch_size)
                X_mini = X[minibatch_index]
                y_mini = y[minibatch_index]
            
            # Forward / Backward Propagation
            train_cost = self.forward_propagation(X_mini, y_mini, lambda_reg)
            grad_thetas, grad_bias = self.backward_propagation(X_mini, y_mini, lambda_reg)
            
            test_cost = self.forward_propagation(X_test, y_test, lambda_reg)
            
            # Append Cost to Cost History
            train_costs.append(train_cost)
            test_costs.append(test_cost)
        
            # Break Loop if Training Costs Approaches Infinity
            if train_cost == np.inf:
                break
            
            # Update Parameters
            if update_method == 'sgd': # Stochastic Gradient Descent
                self.thetas['Theta1'] += -alpha_learn * grad_thetas['Theta1']
                self.thetas['Theta2'] += -alpha_learn * grad_thetas['Theta2']

                self.bias['B1'] += -alpha_learn * grad_bias['B1']
                self.bias['B2'] += -alpha_learn * grad_bias['B2']
            
            elif update_method == 'momentum': # Momentum 
                # Initialize Momentum Terms as Zero
                if len(velocity) == 0:
                    velocity['Theta1'] = np.zeros(grad_thetas['Theta1'].shape)
                    velocity['Theta2'] = np.zeros(grad_thetas['Theta2'].shape)
                    velocity['B1'] = np.zeros(grad_bias['B1'].shape)
                    velocity['B2'] = np.zeros(grad_bias['B2'].shape)
                    
                # Update Velocities
                velocity['Theta1'] = momentum * velocity['Theta1'] - alpha_learn * grad_thetas['Theta1']
                velocity['Theta2'] = momentum * velocity['Theta2'] - alpha_learn * grad_thetas['Theta2']
                
                velocity['B1'] = momentum * velocity['B1'] - alpha_learn * grad_bias['B1']
                velocity['B2'] = momentum * velocity['B2'] - alpha_learn * grad_bias['B2']
                    
                # Update Parameters    
                self.thetas['Theta1'] += velocity['Theta1']
                self.thetas['Theta2'] += velocity['Theta2']

                self.bias['B1'] += velocity['B1']
                self.bias['B2'] += velocity['B2']
                
            # Every Epoch Check Accuracy & Decay Learning Rate
            if iteration == 0 or (iteration+1) % iterations_per_epoch == 0:
                # Check Accuracy
                acc_train = np.mean(self.predict(X_mini) == y_mini)
                acc_test = np.mean(self.predict(X_test) == y_test)
                training_accuracys.append(acc_train)
                test_accuracys.append(acc_test)
                
                # Decay Learning Rate
                if iteration != 0:
                    alpha_learn *= alpha_decay
                
                # Print Costs
                if print_costs:
                    epoch = (iteration+1) // iterations_per_epoch
                    print('Epoch {} of {}: Train Cost={}, Train_Acc={}, Test_Acc={}, Alpha_Learn={}'\
                          .format(epoch,num_epoch,train_cost,acc_train,acc_test,alpha_learn))
                
                # Keep Parameters if Accuracy is Best
                if acc_test > best_accuracy:
                    best_accuracy = acc_test
                    best_parameters['Theta1'] = self.thetas['Theta1']
                    best_parameters['Theta2'] = self.thetas['Theta2']
                    best_parameters['B1'] = self.bias['B1']
                    best_parameters['B2'] = self.bias['B2']
                
                    
        return {'Train Costs':train_costs, 
                'Test Costs':test_costs, 
                'Training Accuracy':training_accuracys,
                'Test Accuracy':test_accuracys,
                'Best Parameters':best_parameters, 
                'Best Accuracy':best_accuracy}
    
                
    def predict(self, X_test):
        # Retrieve Parameters
        theta1, bias1 = self.thetas['Theta1'], self.bias['B1']
        theta2, bias2 = self.thetas['Theta2'], self.bias['B2']
        
        # Layer 1
        a1 = np.copy(X_test)
        
        # Layer 2
        z2 = np.dot(a1, theta1) + bias1
        a2 = self.activation(z2)
        
        # Layer 3 (Output Layer)
        z3 = np.dot(a2, theta2) + bias2
        y_pred = np.argmax(z3, axis=1)
        
        return y_pred
```

## Plot Image & Class Probability

```python
params = results_2['Best Parameters']
def find_prob(X_test, parameters):
    theta1, theta2 = parameters['Theta1'], parameters['Theta2']
    bias1, bias2 = parameters['B1'], parameters['B2']
    
    X_test = np.array(X_test)
    try:
        m, n = X_test.shape
    except:
        m = X_test.shape
    
    # Layer 1
    a1 = np.copy(X_test)
        
    # Layer 2
    z2 = np.dot(a1, theta1) + bias1
    a2 = relu(z2)
        
    # Layer 3 (Output Layer)
    z3 = np.dot(a2, theta2) + bias2
    y_pred = np.argmax(z3, axis=1)
    # Probability
    prob = softmax(z3, m)
    
    return {'Predictions': y_pred, 
            'Probability':prob}
p = find_prob(X_test, params)
```

```python
test_indices = list(y_test.index)
```

```python
def plot_image_prob(p, index):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
    ax1.imshow(rbf_sorted[test_indices[index]])
    ax1.axis('off')
    
    ax2.barh(np.arange(10)+.5 , p['Probability'][index], align='center', height=1,
             color='#5bbad6')
    
    # Prettify
    ax2.set_yticks(np.arange(10)+.5 );
    ax2.set_yticklabels(classes)
    ax2.set_xlim(left=0, right=1)
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.spines['bottom'].set_visible(False)
    ax2.spines['left'].set_visible(False)
    ax2.yaxis.set_ticks_position('left')
    ax2.xaxis.set_ticks_position('bottom')
    
    plt.tight_layout()
```

```python
plot_image_prob(p, 685)
```
