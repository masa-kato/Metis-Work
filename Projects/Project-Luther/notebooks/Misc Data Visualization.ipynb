# Plot Player Salary vs Fantasy Points per Game

```python
rc = {'figure.facecolor': 'black',
      'axes.facecolor': 'white',
      'grid.color': '#FFA500',
      'text.color': 'white',
      'ytick.color': 'white',
      'xtick.color': 'white',
      'axes.labelcolor': 'white'}
```

```python
def seaborn_salary_vs_fppg(df,rc):
    ax = sns.regplot(x='FPPG', y='Salary', data=df, scatter=True, color='black',
                     scatter_kws={'s':80,'facecolors': 'orange',
                                  'edgecolors':'black','linewidths': 2.5})
    
    # Prettify
    plt.rcParams['figure.figsize']=(12,8) 
    sns.plt.title('FPPG vs Salary', fontsize=30)
    sns.plt.xlabel('Fantasy Points per Game', fontsize=24)
    sns.plt.ylabel('Salary', fontsize=24)
    plt.tick_params(axis='both', labelsize=18)
    sns.set_style(style='whitegrid',rc=rc)
```

```python
seaborn_salary_vs_fppg(df,rc)
```
