## Create NMF Model

```python
def create_nmf_model(sparse_matrix, num_topics, lambda_reg=0.1, l1_ratio=0.5):
    '''
    Creates an NMF model
    Input: sparse_matrix - Scipy Sparse Matrix 
           num_topics - Int 
           lambda_reg - Float; Regularization Term
    Output: nmf - Sklearn NMF model
            nmf_topicspace - Numpy Array (Sparse matrix transformed into topic space)
    '''
    nmf = NMF(n_components=num_topics, 
              alpha=lambda_reg, 
              l1_ratio=l1_ratio,
              random_state=0)
    nmf.fit(sparse_matrix)
    nmf_topicspace = nmf.transform(tfidf_sparse)
    return nmf, nmf_topicspace
```

**Let's take a look at the top words for each topic**

```python
def print_topics(vectorizer, nmf_model, num_words):
    '''
    Prints the top words for each topic created by the nmf model
    Input: vectorizer - Sklearn Vectorizer 
           nmf_model - Sklearn NMF model
           num_words - Int     
    Output: None
    '''
    word_list = vectorizer.get_feature_names()
    components = nmf_model.components_
    for i in range(len(components)):
        top_words_index = components[i].argsort()[::-1][:num_words]
        top_words = [word_list[index] for index in top_words_index]
        
        print('Topic {}'.format(i+1))
        print(top_words)
```

```python
print_topics(tfidf_vectorizer, nmf_model_tfidf, num_words=10)
```

```python
def create_scaled_matrix(matrix):
    '''
    Scales matrix using StandardScaler
    Input: Numpy Array
    Output: Numpy Array
    '''
    scaler = StandardScaler()
    return scaler.fit_transform(matrix)
```

```python
def print_top_documents_per_topic(nmf_topicspace, num_documents, df):
    '''
    Prints the top documents for each topic created by the nmf model
    Input: nmf_topicspace - Numpy Array (Matrix of tfidf vector transformed to topic space)
           num_documents - Int (Number of documents to show per topic)   
           df - Pandas DataFrame
    Output: None    
    '''
    index_per_topic = nmf_topicspace.transpose().argsort(axis=1)
    
    for i, topic in enumerate(index_per_topic):
        top_indices = topic[-num_documents:][::-1]
        print('\nTopic {}'.format(i+1))
        for index in top_indices:
            title = df['title'][index]
            score = nmf_topicspace[index].max()
            print('{}, Score: {:.4f}'.format(title, score))
```

## Visualize with t-SNE

```python
def create_tsne_model(topic_space_matrix, dimensions, random_state=0, metric='euclidean', perplexity=30):
    '''
    Creates t-SNE model 
    Input: topic_space_matrix - Numpy Array
           dimensions - Int (2 or 3)
           random_state - Int
           metric - Str (Sklearn pairwise distance metric)
    Output: Sklearn t-SNE Model
            Numpy Array (Embedding of the training data in low-dimensional space)
    '''
    model = TSNE(n_components=dimensions, metric=metric, random_state=random_state, perplexity=perplexity, n_iter=5000)
    tsne_matrix = model.fit_transform(topic_space_matrix)
    return model, tsne_matrix
```

**Convert t-SNE numpy array to Pandas Dataframe**

```python
def create_tsne_df_2D(tsne_matrix, topic_space_matrix, df):
    '''
    Creates a Pandas DataFrame with Columns X, Y, Year, & Topic
    Input: tsne_matrix - Numpy Array
           topic_space_matrix - Numpy Array
           df - Pandas DataFrame (Original DataFrame)
    Output: Pandas DataFrame
    '''
    # Put matrix into a Pandas DataFrame
    tsne_df = pd.DataFrame(tsne_matrix, columns=['X','Y'])
    tsne_df['Year'] = df['year']
    tsne_df['Topic'] = topic_space_matrix.argmax(axis=1)    
    return tsne_df
```

```python
def create_tsne_df_3D(tsne_matrix, topic_space_matrix, df):
    '''
    Creates a Pandas DataFrame with Columns X, Y, Year, & Topic
    Input: tsne_matrix - Numpy Array
           topic_space_matrix - Numpy Array
           df - Pandas DataFrame (Original DataFrame)
    Output: Pandas DataFrame
    '''
    # Put matrix into a Pandas DataFrame
    tsne_df = pd.DataFrame(tsne_matrix, columns=['X','Y','Z'])
    tsne_df['Year'] = df['year']
    tsne_df['Topic'] = topic_space_matrix.argmax(axis=1)    
    return tsne_df
```

```python
def create_tsne_plot_2d(tsne_df, year, topics):
    '''
    Creates a 2D plot of the tsne matrix
    Input: tsne_df - Pandas DataFrame
           year - Int
    Output: None
    '''
    
    data = tsne_df[tsne_df['Year'] <= year]
    
    # Create Plot
    plt.figure(figsize=(10,10))
    colors = plt.cm.Spectral(np.linspace(0, 1, 10))
    patches = []
    
    for i in range(10):
        data_temp = data[data['Topic'] == i]
        color = colors[i]
        plt.scatter(data=data_temp, x='X', y='Y', s=50, c=color)
        patches.append(mpatches.Patch(color=color, label= topics[i]))
        
    
    # Prettify
    plt.axis('off')
    plt.legend(handles=patches, ncol=3, bbox_to_anchor=(1.15, 0.05), fontsize=15)
    plt.title('t-SNE Clusters by Topic', fontsize=30)
    plt.suptitle('NIPS Papers published from 1987 to {}'.format(str(year)), fontsize=20, y=0.88)
```

```python
create_tsne_plot_2d(tsne_df_2d_tfidf, 2016, topics)
```

```python
def create_multiple_tsne_plot_2d(tsne_df, year_one, year_two, topics):
    '''
    Creates a 2D plot of the tsne matrix
    Input: tsne_df - Pandas DataFrame
           year_one - Int
           year_two - Int
    Output: None
    '''
    
    data_one = tsne_df[tsne_df['Year'] <= year_one]
    data_two = tsne_df[tsne_df['Year'] <= year_two]
    
    # Create Figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
    colors = plt.cm.Spectral(np.linspace(0, 1, 10))
    patches = []
    
    for i in range(10):
        data_temp = data_one[data_one['Topic'] == i]
        color = colors[i]
        ax1.scatter(data=data_temp, x='X', y='Y', s=50, c=color)
        patches.append(mpatches.Patch(color=color, label= topics[i]))    
    for i in range(10):
        data_temp = data_two[data_two['Topic'] == i]
        color = colors[i]
        ax2.scatter(data=data_temp, x='X', y='Y', s=50, c=color)
        
    # Prettify
    ax1.axis('off')
    ax2.axis('off')
    plt.legend(handles=patches, ncol=3, bbox_to_anchor=(0.6, 0.05), fontsize=15)
    plt.title('t-SNE Clusters by Topic', fontsize=30, x=-0.1)
    plt.figtext(0.17,0.87,'NIPS Papers published from 1987 to {}'.format(str(year_one)), fontsize=20)
    plt.figtext(0.58,0.87,'NIPS Papers published from 1987 to {}'.format(str(year_two)), fontsize=20)
```

```python
create_multiple_tsne_plot_2d(tsne_df_2d_tfidf, 2000, 2016, topics)
```

**Now let's try to plot how many papers were published on each topic every year**

```python
def create_paper_counts_per_topic_per_year(tsne_df):
    '''
    Creates a list of document counts for each topic per year
    Input: tsne_df - Pandas DataFrame
    Output: List of lists
    '''
    years = list(set(tsne_df['Year']))
    topics = list(set(tsne_df['Topic']))
    
    documents_per_topic_per_year = []
    for topic in topics:
        counts = []
        topic_df = tsne_df[tsne_df['Topic'] == topic]
        for year in years:
            num_documents = len(topic_df[topic_df['Year'] == year])
            counts.append(num_documents)
        documents_per_topic_per_year.append(counts)
    return documents_per_topic_per_year
```

```python
paper_counts_per_topic = create_paper_counts_per_topic_per_year(tsne_df_2d_tfidf)
years = list(set(tsne_df_2d_tfidf['Year']))
```

```python
def plot_paper_counts_per_topic_per_year(paper_counts_per_topic, topics, years, highlight=None):
    '''
    Plots the number of papers for each topic per year
    Input: paper_counts_per_topic - list of lists
           topics - list of topic names
           years - list of ints
           highlight - int
    Output: None
    '''    
    fig, ax = plt.subplots(figsize=(16,8))
    colors = plt.cm.Spectral(np.linspace(0, 1, 10))
        
    for index, counts in enumerate(paper_counts_per_topic):
        if type(highlight) == int:
            if index == highlight:
                alpha = 1
            else: alpha = 0.05
        else: alpha = 1
        plt.plot(years, counts, label=topics[index], c=colors[index],linewidth=3,alpha=alpha)
    # Prettify
    sns.set_style('white')
#     plt.title('Published Papers by Year per Topic', fontsize=20)
#     plt.xlabel('Year', fontsize=20)
#     plt.ylabel('Number of Papers', fontsize=20)
    plt.xlabel('', fontsize=20)
    plt.ylabel('', fontsize=20)
    
    plt.xticks(fontsize=14,rotation='vertical')
    plt.yticks(fontsize=14)
    plt.xlim([min(years),max(years)])
    plt.legend(bbox_to_anchor=(1.02,-0.15), ncol=4, fontsize=16)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    
#     plt.savefig('../images/num_papers_per_topic_line_plot_nolabels.png', transparent=True)
```

```python
plot_paper_counts_per_topic_per_year(paper_counts_per_topic, topics, years, highlight=4)
```
